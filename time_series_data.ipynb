### Bonus: Visualize the daily min, average, and max temperature over time on a single line plot, i.e. the min, average, and maximum temperature should be 3 seperate lines.
df = data.sf_temps()

```python
df
```

```python
# plt.figure(figsize=(14,10))

df.temp.plot(label='daily')
df.resample('D').min().temp.plot(label='min')
df.resample('D').mean().temp.plot(label='average')
df.resample('D').max().temp.plot(label='max')

plt.title('plotting close over time with various resampling techniques')
plt.legend()
plt.show()
```
