```python
response = requests.get('https://swapi.dev/api/people/')
data = response.json()
df = pd.DataFrame(data['results'])

while data['next'] != None:
    response = requests.get(data['next'])
    data = response.json()
    print(data['next'])
    df = pd.concat([df, pd.DataFrame(data['results'])]).reset_index(drop=True)
```
